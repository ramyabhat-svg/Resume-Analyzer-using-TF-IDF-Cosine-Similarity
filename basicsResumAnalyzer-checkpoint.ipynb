```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
    "Python is great for data science and machine learning.",
    "Java is commonly used for large enterprise applications.",
    "Python developers use Python for web development and automation."]
v= TfidfVectorizer()
transformed=v.fit_transform(corpus)
print(v.vocabulary_ )
```

```python
print(transformed)
```

```python
print(transformed.toarray()) 
```

```python
allnames=v.get_feature_names_out()
for word in allnames:
    indx=v.vocabulary_.get(word)
    print(f"{word}: {v.idf_[indx]}")
```
