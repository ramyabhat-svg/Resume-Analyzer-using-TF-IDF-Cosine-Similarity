```python
from sklearn.feature_extraction.text import TfidfVectorizer
import PyPDF2 
from sklearn.metrics.pairwise import cosine_similarity, cosine_distances

job_d="Participated in technical workshops, hackathons, and community events focused on Google technologies.Collaborated with peers on projects involving cloud computing, web development, and machine learning.Contributed to problem-solving sessions and helped organize student-led tech activities.Engaged in learning pathways such as Firebase, Android Development, and Google Cloud programs."

b=PyPDF2.PdfReader("Resume.pdf") 
text="" 
for i in b.pages: 
    text+=i.extract_text() 

docs = [job_d, text]
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(docs)

similarity = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:2])[0][0]

print("the resume matches", round(similarity*100,2), "%")

skills_list = [
    "python", "sql", "excel", "tableau", "power bi", 
    "machine learning", "pandas", "numpy", "scikit-learn",
    "git", "data analysis", "data visualization"
]

def extract_skills(text, skills):
    text_lower = text.lower()
    found = []
    for skill in skills:
        if skill in text_lower:
            found.append(skill)
    return found

jd_skills = extract_skills(job_d, skills_list)
resume_skills = extract_skills(text, skills_list)

missing_skills = [skill for skill in jd_skills if skill not in resume_skills]

print("\nSkills required by Job:", jd_skills)
print("Skills found in Resume:", resume_skills)
print("Missing Skills:", missing_skills)
```
